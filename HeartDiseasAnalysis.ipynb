Data Preprocessing

```python
# Feature Selection: Dropping less correlated features
selected_features = ["age", "sex", "cp", "thalach", "oldpeak", "ca", "thal"]
X = data[selected_features]
y = data["target"]
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

1. Classification 

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score


```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Hyperparameter Tuning for Random Forest

param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
rf = RandomForestClassifier(random_state=42)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='accuracy')
grid_rf.fit(X_train, y_train)
print(f"Best parameters for Random Forest: {grid_rf.best_params_}")
rf_best = grid_rf.best_estimator_
```

```python
# Train and evaluate the optimized Random Forest
rf_best.fit(X_train, y_train)
y_pred = rf_best.predict(X_test)
```

```python
# Classification Report
print("\nClassification Report for Random Forest:")
print(classification_report(y_test, y_pred))
```

```python
results = [{
    "Model": "Random Forest",
    "Accuracy": accuracy_score(y_test, y_pred),
    "Precision": precision_score(y_test, y_pred),
    "Recall": recall_score(y_test, y_pred),
    "F1 Score": f1_score(y_test, y_pred)
}]
```

```python
# Convert results to a DataFrame and display as a table
results_df = pd.DataFrame(results)
print("\nClassification Results:")
print(results_df)
```

```python
# Visualize classification results
plt.figure(figsize=(8, 6))
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
values = results_df.loc[0, metrics]
plt.bar(metrics, values, color=['#FFA07A', '#4682B4', '#66CDAA', '#FF6347'])
plt.title("Random Forest Classification Metrics")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```

2. Clustering 

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)
data["Cluster"] = kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score
# Evaluate clustering
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print(f"\nSilhouette Score for Clustering: {silhouette_avg}")
```

```python
# Elbow Method to Determine Optimal Clusters
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 10), inertia, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# Visualize Clusters

plt.figure(figsize=(10, 6))
plt.scatter(data["age"], data["chol"], c=data["Cluster"], cmap="viridis")
plt.title("Clustering Visualization")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.colorbar(label="Cluster")
plt.show()
```
